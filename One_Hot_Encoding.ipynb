```python
#One Hot Encoding
import pandas as pd

infile1 = r"C:\Users\Bhavyashree\Desktop\ML\OneHotEncoding\HomePrices.csv"
df1 = pd.read_csv(infile1)
df2 = pd.get_dummies(df1.town)
print(df1)
print(df2)
```

```python
df3 = pd.concat([df1,df2], axis = 'columns')
df3
```

```python
df3 = df3.drop(['town', 'west windsor'], axis = 'columns')
df3
```

```python
y = df3['price']
y
```

```python
x = df3.drop(['price'], axis = 'columns')
x
```

```python
from sklearn import linear_model

lin_reg_obj = linear_model.LinearRegression()
lin_reg_obj.fit(x,y)
```

```python
lin_reg_obj.predict([[3500,1,0]])
```

```python
lin_reg_obj.score(x,y)
```

```python
lin_reg_obj.predict([[2800,0,1]])
```
